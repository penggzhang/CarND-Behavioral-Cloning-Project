```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.axes_grid1 import ImageGrid
import cv2

from keras.models import Sequential
from keras.layers import Input, Lambda, Convolution2D, MaxPooling2D
from keras.layers import Activation, Flatten, Dense, Dropout
from keras.regularizers import l2
from keras.optimizers import Adam
from keras.models import model_from_json

import os, json
from pathlib import Path
```

**Load driving log, preview data**

```python
# Load driving log.
# Helper function for load driving_log.csv into a data frame.
def load_log(log_fpath, header):
    driving_log_file = log_fpath
    log_df = pd.read_csv(driving_log_file, 
                         header=header,
                         names=['center','left','right','steer','throttle','break', 'speed'],
                         index_col = False)
    # Give data summary
    print("Shape of data frame: {}".format(log_df.shape))
    return log_df
# Note: Pass header = 0 for csv with header, 
# and None without header.
```

```python
log_fpath = "data/driving_log_all.csv"
header = 0
log_df = load_log(log_fpath, header)
```

```python
# Preview steerings.
# Helper function for plotting steerings-samples.
def plot_steerings(df, field='steer', k=300):
    x = np.arange(k)
    y = np.array(df[field][:k])
    
    plt.plot(x, y, 'b-')
    plt.title("Preview {} steerings".format(k))
    plt.xlabel("Sample")
    plt.ylabel("Steering")
    plt.ylim(-1, 1)
    plt.show()
```

 **Filter samples**

```python
# Helper function for removing samples with throttle below threshold.
def filter_samples(log_df, throttle_threshold):
    log_df = log_df[log_df['throttle'] > throttle_threshold]

    # Reset the index of filtered data frame.
    log_df.reset_index(drop=True)

    nb_samples = log_df.shape[0]
    print("\nNumber of samples after filtering: {}".format(nb_samples))
    return log_df, nb_samples
```

```python
throttle_threshold = 0.1
log_df, nb_samples = filter_samples(log_df, throttle_threshold)
```

**Shuffle samples by log**

```python
# Helper function for shuffling samples by log
def shuffle_samples(log_df):
    log_df = log_df.sample(frac=1).reset_index(drop=True)
    print("\nSamples shuffled.")
    return log_df
```

```python
log_df = shuffle_samples(log_df)

# Preview steerings again.
plot_steerings(log_df)
```

**Split train and validation sets by log**

```python
# Helper function for splitting train and validation sets
def split_train_and_valid(log_df, nb_samples, valid_split=0.2):
    # Generate a Boolean mask for splitting.
    msk = (np.random.rand(nb_samples) < valid_split)

    # Split with the mask and reset index.
    valid_log = log_df[msk].reset_index(drop=True)
    train_log = log_df[~msk].reset_index(drop=True)

    nb_samples_train = train_log.shape[0]
    nb_samples_valid = valid_log.shape[0]
    assert nb_samples_train + nb_samples_valid == nb_samples
    print("\nNumbers of train and validation samples: {0}, {1}".\
          format(nb_samples_train, nb_samples_valid))

    return train_log, valid_log, nb_samples_train, nb_samples_valid
```

```python
valid_split = 0.05
train_log, valid_log, nb_samples_train, nb_samples_valid = \
    split_train_and_valid(log_df, nb_samples, valid_split)
```

```python
# Helper function for modifying path strings in order to 
# make them as corret local paths. This step is for 
# implementation of this program on other places, like AWS.
def modify_img_fpath(df, cameras, img_parent_dir, delete):
    n = df.shape[0]
    if delete:
        for c in cameras:
            for i in range(n):
                # Find substring prior to local path.
                ind = df[c][i].find(img_parent_dir)
                substring = df[c][i][:ind]

                # Delete the substring.
                new_path = df[c][i].replace(substring, '')
                df.set_value(i, c, new_path.strip())
    else:
        for c in cameras:
            for i in range(n):
                # Add local_path substring to entries.
                new_path = img_parent_dir + '/' + df[c][i].strip()
                df.set_value(i, c, new_path)
                
    # Check modified path strings.
    print("\nFile path of 1st sample:")
    for c in cameras:
        print(df[c][0])
    print()    
    return df
```

**Preview images**

```python
# Helper function for loading image.
def load_image(fpath):
    image = cv2.imread(fpath)
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
# Note: cv2 load image as BGR, so convert back to RGB.
```

```python
# Helper function for plotting single image
# with camera as x-label, and steer as title.
def plot_ax(ax, img, cam, steer):
    ax.imshow(img)
    ax.set_title("steer: {:.5f}".format(steer), fontsize=15)
    ax.set_xlabel(cam, fontsize=18)
    for sp in ax.spines:
        ax.spines[sp].set_visible(False)
```

```python
# Helper function for previewing left, center, right images
# for 5 training samples.
def plot_preview(log, d, preprocess):
    # Random pick 5 training samples by indexing.
    ind = np.random.choice(train_log.index, 5)

    # For the picked samples, collect images along with their
    # camera and steer information into a list of tuples.
    if preprocess:
        images = [( preprocess_image(train_log[c][i]), c, train_log[d[c]][i] )
                  for i in ind for c in cameras]
    else:
        images = [( load_image(train_log[c][i]), c, train_log[d[c]][i] )
                  for i in ind for c in cameras]
        
    print("Shape of image: {}".format(images[0][0].shape))

    fig = plt.figure(figsize=(20, 15))
    grid = ImageGrid(fig, 111, nrows_ncols=(5, 3),
                     axes_pad=0.5, label_mode='L')
    for i, g in enumerate(grid):
        plot_ax(g, *images[i])
    plt.show()
```

```python
# Dictionary for correponding different images to steers.
d = {'left'  : 'steer', 
     'center': 'steer',
     'right' : 'steer'}

# Preview original images
plot_preview(train_log, d, False)
```

**Preprocess image**

```python
# Helper function for preprocessing image.
# 1) Remove the hood from image. Crop top 50 pixels 
#    and bottom 30 pixels from each image.
# 2) Then downsize each image by a factor of 2
#    on either dimension. 
def preprocess_image(fpath):
    image = load_image(fpath)
    row, col = image.shape[0], image.shape[1]
    
    # Crop top and bottom pixels.
    top_crop = 50
    btm_crop = 30
    image = image[top_crop:(row - btm_crop), :]
    
    # Resize to new row and col.
    downsize = 2
    new_row = (row - top_crop - btm_crop) // downsize
    new_col = col // downsize
    image = cv2.resize(image, 
                       (new_col, new_row), 
                       interpolation=cv2.INTER_AREA)
        
    return image 
```

```python
# Preview preprocessed images
plot_preview(train_log, d, True)
```

**Generate steers for left and right images**

```python
# Helper function for incorporating left and right images.
# After trial and error, add a steer of 0.25 for the left image, 
# which means we have to steer further right back to center. 
# Also subtract a steer of 0.25 for right image, which means
# we have to steer further left back to center.
def gen_left_and_right_steers(log, steer_tune):
    base = np.array(log['steer'])
    log = log.assign(left_steer=pd.Series(base+steer_tune).values)
    log = log.assign(right_steer=pd.Series(base-steer_tune).values)
    assert len(log['left_steer']) == len(log['right_steer']) == len(log['steer'])
    print("\nSteers were generated for left and right images.")
    return log
```

```python
# Generate steers for left and right images for training set.
steer_tune = 0.05
train_log = gen_left_and_right_steers(train_log, steer_tune)
```

**Create keras fit_generator**

```python
# Helper function for generating data batch by batch.
# Each batch contains a tuple of batch images and steers.
def generate_data(log, cameras, d, batch_size):
    n = log.shape[0]
    while 1: 
        for offset in range(0, n, batch_size):
            batch_images, batch_steers = [], []
            k = min(n - offset, batch_size)
            for i in range(k):
                row = offset + i
                for c in cameras:
                    fpath = log[c][row]
                    img = preprocess_image(fpath)
                    steer = log[d[c]][row]
                    
                    batch_images.append(img)
                    batch_steers.append(steer)
                    
                    # Mirror images and label with negative steers
                    batch_images.append(cv2.flip(img, 1))
                    batch_steers.append(-float(steer))
                    
            batch_images = np.array(batch_images)
            batch_steers = np.array(batch_steers)
            yield batch_images, batch_steers
```

```python
# Check training data generator.
cameras = ['center']
d_with_LR = {'center': 'steer'}
batch_size = 128
g_valid = generate_data(valid_log, cameras, d_with_LR, batch_size)
tmp = next(g_valid)

print("Shape of one batch of image data: {}".format(tmp[0].shape))
print("Shape of one batch of steer data: {}".format(tmp[1].shape))
```

**Design and create model**

Use Nvidia architecture as the baseline model.

Add a 3@1x1 convolutional layer after the 'Normalization'. That'll allow the networks to learn its best color space.
It is then followed by 5 convolutional layers, which are 24@3x3, 36@3x3, 48@3x3, 64@2x2 and 64@1x1 filters respectively. Each of these layers will be downsampled by max pooling.

Then flattened features will go through 3 fully connected layers with neurons of 100, 50, 10 respectively.

```python
# Helper function for model design
def create_model(input_shape):
    model = Sequential()

    # Normalization
    model.add(Lambda(lambda x: x / 127.5 - 1.0, input_shape=input_shape))

    # 1x1 convolutional layer
    model.add(Convolution2D(3, 1, 1, border_mode='valid', name='conv0'))
    model.add(Activation('relu'))

    model.add(Convolution2D(24, 3, 3, border_mode='valid', name='conv1'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Activation('relu'))

    model.add(Convolution2D(36, 3, 3, border_mode='valid', name='conv2'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Activation('relu'))

    model.add(Convolution2D(48, 3, 3, border_mode='valid', name='conv3'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Activation('relu'))

    model.add(Convolution2D(64, 2, 2, border_mode='valid', name='conv4'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Activation('relu'))

    model.add(Convolution2D(64, 1, 1, border_mode='valid', name='conv5'))
    model.add(MaxPooling2D(pool_size=(1, 2)))
    model.add(Activation('relu'))

    model.add(Dropout(0.5))

    model.add(Flatten())

    model.add(Dense(100, W_regularizer=l2(5e-4), name='fc1'))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(50, W_regularizer=l2(5e-4), name='fc2'))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(10, W_regularizer=l2(5e-4), name='fc3'))
    model.add(Activation('relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1, name='output'))
    
    return model
```

```python
# Create a model
input_shape = (40, 160, 3)
model = create_model(input_shape)

# Model summary
model.summary()
```

**Compile model**

```python
# Helper function for compiling the model.
# Fine tune the learning rate, lr, along training sessions.
def compile_model(model, lr):
    # Optimizer: ADAM
    adam = Adam(lr=lr)
    # Optimization objective: mean square error.
    model.compile(optimizer=adam, loss='mse')
    return model
```

**Save model**

```python
# Helper function for saving model.
def save_model(model, json_file, weights_file):
    json_string = model.to_json()
    
    # Save architecture to JSON file.
    if Path(json_file).is_file():
        os.remove(json_file)
    with open(json_file, 'w') as f:
        json.dump(json_string, f)

    # Save weights to HDF5 file.
    if Path(weights_file).is_file():
        os.remove(weights_file)
    model.save_weights(weights_file)
    
    print("Model saved.")
```

**Train model**

```python
# Helper function for training model.
def train(nb_epochs, save, model_id):
    global model, cameras, d, d_train, train_log, valid_log
    global nb_samples_train, nb_samples_valid, batch_size
    
    # 1: the mirror sample
    nb_samples_per_epoch = (len(cameras) + 1) * nb_samples_train
    
    for i in range(nb_epochs):

        # Create generators for train and validation data.
        train_generator = generate_data(train_log, cameras, d_train, batch_size)
        valid_generator = generate_data(valid_log, ['center'], d, batch_size)

        # Train model on data batch-by-batch.
        # Also validate on validation data batch-by-batch.
        history = model.fit_generator(generator=train_generator,
                                      samples_per_epoch=nb_samples_per_epoch,
                                      nb_epoch=1,
                                      validation_data=valid_generator,
                                      nb_val_samples=nb_samples_valid)

        if save:
            # Save models for each training epoch.
            json_file = 'models/model_' + str(model_id) + '.json'
            weights_file = 'models/model_' + str(model_id) + '.h5'
            save_model(model, json_file, weights_file)
            model_id += 1
            # Note: Make directory 'models' beforehand.
            
    return model
```

**Load pre-trained model**

```python
# Helper function for loading a pre-trained model
def load_model(json_file, weights_file):
    with open(json_file, 'r') as f:
        model = model_from_json(json.loads(f.read()))
    model.load_weights(weights_file, by_name=True)
    print("Model loaded.")
    return model  

# Note: Need to change the line in drive.py as below:
#    model = model_from_json(json.loads(jfile.read()))
```
